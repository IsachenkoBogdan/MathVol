# 1.1 Рекуррентные соотношения

Про рекурсию в python (кому интересно)

https://pythontutor.ru/lessons/functions/

https://foxford.ru/wiki/informatika/rekursiya-v-python

````python
def factorial(n):
    res = 1
    for i in range(1, n + 1):
        res *= i
    return res


print(factorial(3))
print(factorial(51))
````

````python
### Рекурсивное вычисление факториала

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)


print(factorial(3))
print(factorial(51))
````

### Task 1.2.2 Добавить на график task 1.2.1 точки персечения графиков СИНИМ цветом и точки пересечения функций с осью OX фиолетовым.



````python
### Task 1.2.2. Solution:
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure()
x = np.arange(-12, 12, 0.1)
plt.plot(x, 7*x*x*x + 3*x*x - x - 9)
plt.plot(x, 2*x*x - 7*x + 3)
plt.plot(x,15 - 1.5*x)
plt.scatter(1 , 1 , color='magenta')
from scipy.optimize import fsolve
def func1(x):
    return 7*x*x*x + 3*x*x - x - 9
plt.scatter(fsolve(func1,0.01), 0 , color='magenta')
def func2(x):
    return 2*x*x - 7*x + 3
plt.scatter(fsolve(func2,0.01), 0 , color='magenta')
def func3(x):
    return 15 - 1.5*x
plt.scatter(fsolve(func3,0.01), 0 , color='magenta')
plt.grid()
def func4(x):
    return func1(x)-func2(x)
plt.scatter(fsolve(func4,0),func2(fsolve(func4,0)), color='blue')
def func5(x):
    return func1(x)-func3(x)
plt.scatter(fsolve(func5,0),func3(fsolve(func5,0)), color='blue')
def func6(x):
    return func2(x)-func3(x)
plt.scatter(fsolve(func6,0),func2(fsolve(func6,0)), color='blue')
plt.scatter(4.184,func3(4.184),color='blue')
plt.ylim(-100,100)
plt.show()
````

````python
 # Получение данных из удаленного репозитория
    
import requests  # pip install requests

url = u'https://raw.githubusercontent.com/pensnarik/russian-cities/master/russian-cities.json'

json_data = requests.get(url).json()

print(json_data)

````

````python
# Example Task 1.5

your_subject = 'Крым'


def get_all_cities_of_this_subject(subject):
    data_s = []
    for item in json_data:
        if item['subject'] != subject:
            continue
        data_s.append(item)
    return data_s



data = get_all_cities_of_this_subject(your_subject)

# Показ структуры полученных данных
print("\n\n Исходные даннные по субъекту такие: \n")
print(data)


# Подготовка данных для отрисовки

cities = []
population = []

for city in data:
    cities.append(city['name'])
    population.append(city['population'])
    
    
print("\n\n Подготовленные данные по региону: \n")
print(cities, population)
    

````

````python
# Попробуем визуализировать подготовленные данные

plt.bar(cities, population, align='center')
plt.show()

# Попробуем поменять месатми оси

plt.barh(cities, population, align='center')
plt.show()
````

````python
### Task 1.4.1: Solution:
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
your_subject = 'Томская область'
import requests  # pip install requests

url = u'https://raw.githubusercontent.com/pensnarik/russian-cities/master/russian-cities.json'

json_data = requests.get(url).json()


def get_all_cities_of_this_subject(subject):
    data_s = []
    for item in json_data:
        if item['subject'] != subject:
            continue
        data_s.append(item)
    return data_s



data = get_all_cities_of_this_subject(your_subject)

cities = []
population = []

for city in data:
    cities.append(city['name'])
    population.append(city['population'])
    
plt.barh(cities, population, align='center')
plt.show()
````

````python

print(population)


### Найдем среднее значение

sum = 0
for n in population:
    sum += n
    
number = len(population)

print(number)
print(sum)

print("Middle: ", sum/number)
### Вывод в целом int
print("Middle: ", int(sum/number))
````
